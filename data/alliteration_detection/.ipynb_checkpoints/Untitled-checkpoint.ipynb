# Alliteration Detection

Write an algorithm that marks a sentence as a sentence containing an alliteration if all words in a certain concatenation of words in the sentence start with a comparable sound. The "sound" part is important, as sentences like "Der frühe Vogel fängt den Wurm" should be classified as an alliteration, as f and v sound very similar in that particular case. 

```python
def amount_of_words_following_word(index, sentence):
    if(index >= len(sentence) - 1):
        return 0
    if(sentence[index][0] != sentence[index+1][0]):
        return 0
    else:
        return 1 + amount_of_words_following_word(index+1,sentence)
```

```python
sentence = "Physik für Freunde"
```

```python
sentence = sentence.lower().split(" ")
```

```python
amount_of_words_following_word(0,sentence)
```

```python
index = 0
while index < len(sentence):
    print(amount_of_words_following_word(index,sentence))
    index += 1
```

## Introduce the cologne phonetics

```python
import cologne_phonetics
```

```python
# ([^CKQckq][Xx]) Regex which will be used to ignore these cases
cologne_phonetics.encode("X")
```

```python
def first_letter_cp_encoding(sentence_encoding,index):
    return sentence_encoding[index][1][0]

def alliteration_amount(index,cp_sentence):
    if(index >= len(cp_sentence) - 1):
        return 0
    if(first_letter_cp_encoding(cp_sentence,index) != first_letter_cp_encoding(cp_sentence,index+1)):
        return 0
    else:
        return 1 + alliteration_amount(index+1,cp_sentence)
```

```python
word_encoding = cologne_phonetics.encode("Physik für Freunde")
```

```python
alliteration_amount(0,word_encoding)
```

Function to prefilter sentences. Will return a list containing the amount of steps necessary for each word to find a word with the same number at the start

```python
test_1 = "Dieser Satz hat keine Alliteration"
test_2 = "Fischer Fritz angelt frische Fische"
```

```python
def get_preprocessed_sentence(sentence):
    if type(sentence) == str:
        return sentence.lower().split(" ")
    else:
        return [""]

def get_cologne_encoding_as_list(sentence):
    word_encoding = cologne_phonetics.encode(sentence)
    res = []
    
    for word,encoding in word_encoding:
        res.append(encoding)
    return res

def has_word_with_same_letter_at_gap(starting_index, gap, preprocessed_sentence):
    
    step = starting_index + gap
        
    if(step >= len(preprocessed_sentence)):
        return False
    
    if (len(preprocessed_sentence[starting_index]) < 1):
        return False
    
    if (len(preprocessed_sentence[step]) < 1):
        return False

    else:
        return preprocessed_sentence[starting_index][0] == preprocessed_sentence[step][0]

def contains_alliteration(sentence,gap,sentence_preprocessing_func):
    preprocessed_sentence = sentence_preprocessing_func(sentence)
    res = []
    
    index = 0
    
    while index < len(preprocessed_sentence):
        
        res.append(has_word_with_same_letter_at_gap(index,gap,preprocessed_sentence))
        index += 1
    
    return True in res
    
    
```

```python
def has_at_least_one_alliteration_case(sentence):
    case_1 = contains_alliteration(sentence,1,sentence_preprocessing_func=get_preprocessed_sentence)
    case_2 = contains_alliteration(sentence,2,sentence_preprocessing_func=get_preprocessed_sentence)
    case_3 = contains_alliteration(sentence,1,sentence_preprocessing_func=get_cologne_encoding_as_list)
    case_4 = contains_alliteration(sentence,2,sentence_preprocessing_func=get_cologne_encoding_as_list)
    
    return case_1 or case_2 or case_3 or case_4
    
```

```python
for elem in alliterations:
    if not has_at_least_one_alliteration_case(elem):
        print(elem)
```
